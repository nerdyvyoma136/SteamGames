# Data Preprocessing

### Small data files combined in one for faster processing

```python
#First 4 files concated in one big file
qt1= pd.concat([df1,df2,df3,df4], axis=0) #1st big file containing data from 4 smaller files
qt1.head()
```

```python
'''To consider serious and meaning reviews,votes_funny columns should have value 0,hence creating another dataframe to 
keep funny review aside from actual data.'''
qt1_vf0=qt1[(qt1['votes_funny']>0)] 
qt1_vf0.head()
```

```python
#Considering data with serious review votes, hence votes_funny==0
qt1=qt1[(qt1['votes_funny']==0)]
qt1.head()
```

```python
qt1=qt1.drop(['unix_timestamp_created','weighted_vote_score'],axis=1)#Discarding unix_timestamp_created as to consider 
                                                                    #latest updated date and weighted vote score because more 0 values
```

```python
qt1=qt1.dropna(axis=0, how='any')#Remove rows having any column null values in it
qt1.info()
```

```python
qt1.shape
```

```python
from datetime import datetime


qt1['normal_updated_date']= pd.to_datetime(qt1['unix_timestamp_updated'], unit='s')
qt1.head()
```

```python
qt1.drop(['unix_timestamp_updated'], axis=1, inplace=True) #Drp unix format updated date
```

```python
#Split readable date into month, year etc.
qt1['Updated_Year'] = qt1['normal_updated_date'].dt.year
qt1['Updated_Month'] = qt1['normal_updated_date'].dt.month
qt1['Updated_Day'] = qt1['normal_updated_date'].dt.day
qt1['Updated_Weekday'] = qt1['normal_updated_date'].dt.day_name()
qt1['Updated_Hour'] = qt1['normal_updated_date'].dt.hour
qt1['Updated_Minute'] = qt1['normal_updated_date'].dt.minute
```

```python
#To work only for a decade data considering rows having year > 2010
qt1=qt1[(qt1['Updated_Year']>2010)]
qt1.head()
```

```python
#After all operations checking the shape of data
qt1.shape
```

```python
#Storing this preprocessed file to use in next operations instead of everytime running 11 files 
qt1.to_csv(r'Qt1data.csv', index=False)
```
